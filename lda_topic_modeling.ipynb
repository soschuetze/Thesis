```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import pyLDAvis
import pyLDAvis.sklearn
import re
import pandas as pd
import matplotlib.pyplot as plt
import nltk
import string
pyLDAvis.enable_notebook()
```

```python
stopwordsList = nltk.corpus.stopwords.words('english')
punctuation = string.punctuation
```

```python
tweets_df = pd.read_csv("tweets_cleaned.csv", index_col = [0])
tweets_df = tweets_df.dropna()
docs_raw = tweets_df['text'].tolist()
```

```python
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 0.5, 
                                min_df = 10)

dtm_tf = tf_vectorizer.fit_transform(docs_raw)
print(dtm_tf.shape)
```

```python
tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())

dtm_tfidf = tfidf_vectorizer.fit_transform(docs_raw)
print(dtm_tfidf.shape)
```

```python
# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_components=5, random_state=0)
lda_tfidf.fit(dtm_tfidf)
```

```python
pyLDAvis.sklearn.prepare(lda_tfidf, dtm_tfidf, tf_vectorizer)
```

```python
# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_components=10, random_state=0)
lda_tfidf.fit(dtm_tfidf)
```

```python
pyLDAvis.sklearn.prepare(lda_tfidf, dtm_tfidf, tf_vectorizer)
```
